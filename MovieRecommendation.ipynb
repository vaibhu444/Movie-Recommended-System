```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("files/tmdb_5000_movies.csv")
credits = pd.read_csv("files/tmdb_5000_credits.csv")
```

```python
# merge the movies dataframe with credits based on title column in both df 
movies = movies.merge(credits, on="title")
```

```python
movies.head()
```

## need to filter out usefull column from above data set (feature extraction)
1. 'movie_id'
2. 'title'
3. 'overview'
4. 'genres'
5. 'keywords'
6. 'cast'
7. 'crew'

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head()
```

```python
movies.dropna(inplace=True)
```

```python
import ast
# ast library of python is used to convert string like python data structure(dict,list) into python DS 

def extract(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(extract)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(extract)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(extract)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x: x[0:3])
movies.head()
```

```python
import ast
def fetch_director(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job']=="Director":
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head(100)
```

```python
def remove_spaces(data):
    L1=[]
    for d in data:
        L1.append(d.replace(" ",""))
    return L1
```

```python
movies['genres'] = movies['genres'].apply(remove_spaces)
movies['keywords'] = movies['keywords'].apply(remove_spaces)
movies['cast'] = movies['cast'].apply(remove_spaces)
movies['crew'] = movies['crew'].apply(remove_spaces)
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

- through CountVectorizer i am converting raw data into numerical data 

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

cosine_similarity is used to calculate the distance between vectors in this case it will calculate the distance of 4806x4806 movie's distance and cosine_similarity normalize it between 0 to 1 

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
def recommend(movie_name):
    index = new[new['title']==movie_name].index[0]
    similar_movies = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x:x[1])
    for i, s in similar_movies[1:6]:
        print(new.iloc[i].title)
```

```python
recommend('Avatar')
```
